# **Tipos de Variáveis em Python**
Existem vários tipos de variáveis em Python, incluindo:

* Inteiros (int): São números inteiros, positivos ou negativos, sem parte decimal. Exemplo: 5, -10, 0.
* Floats (float): São números decimais, positivos ou negativos. Exemplo: 3.14, -2.5, 0.0.
* Booleanos (bool): São valores binários que representam Verdadeiro (True) ou Falso (False). Exemplo: True, False.
* Strings (str): São sequências de caracteres. Exemplo: "Olá, mundo!", "Python é incrível".
* Listas (list): São coleções ordenadas de itens, que podem ser de tipos diferentes. Exemplo: [1, 2, 3], ["a", "b", "c"].
* Tuplas (tuple): São coleções ordenadas de itens, que podem ser de tipos diferentes. Diferente das listas, as tuplas são imutáveis. Exemplo: (1, 2, 3), ("a", "b", "c").
* Dicionários (dict): São coleções não ordenadas de pares chave-valor. Exemplo: {"nome": "Maria", "idade": 30}.

# **Estruturas de repetição em Python**

Antes de começarmos, é importante entendermos o conceito de iteração. A iteração é uma forma de repetir uma determinada sequência de instruções um certo número de vezes, ou enquanto uma condição for verdadeira. As estruturas de repetição em Python são usadas para implementar a iteração de forma mais eficiente e organizada.

Existem duas principais estruturas de repetição em Python: o '''**for**''' e o '''**while**'''. Vamos explicar cada um deles com exemplos simples de código:

# ***Estrutura de repetição "for"***

O for é uma estrutura de repetição que permite iterar sobre uma sequência de valores, como uma lista, uma string ou um intervalo numérico. A sintaxe básica do for é:

````python
##Estrutura, não rodar...
for valor in sequencia:
    # instruções a serem repetidas
````

Onde valor é a variável que irá assumir cada valor da sequência, e sequencia é a sequência a ser percorrida. As instruções dentro do bloco de código do for serão executadas uma vez para cada valor da sequência.

Aqui está um exemplo de um for que itera sobre uma lista de números e imprime cada valor:

````python
##Exemplo, imprimindo dados de uma lista
numeros = [1,2,3,4,5]

##montar a estrutura de repetição
for numero in numeros:
  print(numero)
````

````python
##Exemplo de soma em uma lista
##montar a estrutura de repetição
for numero in numeros:
  print(numero+1)
````

````python
##Estrutura para acumular
lista =[1,2,3,4,5]
soma= 0

for elemento in lista:
  #+= : é equivalente a fazer x = x + valor . Por exemplo, se x valer 10 e fizermos x += 2 , x passará a ter o valor 12.
  soma += elemento

print(soma)
````

O while é uma estrutura de repetição que permite executar um bloco de código enquanto uma determinada condição for verdadeira. A sintaxe básica do while é:

````python
## Estrutura
while condicao:
    # instruções a serem repetidas
````

````python
### Itera enquanto valor < x
i = 0

## Enquanto i for < x print a o valor de i+= 1
while i < 5:
  print(i)
  #+= : é equivalente a fazer x = x + valor . Por exemplo, se x valer 10 e fizermos x += 2 , x passará a ter o valor 12.
  i += 1

### É importante ter cuidado ao usar o while,
###pois se a condição nunca se tornar falsa, o loop ficará preso em um
````

````python
#Lendo um dataframe
df = pd.read_csv('/content/sample_data/california_housing_test.csv')
df
````

````python
df.columns
````

````python
##Vamos filtrar dados
filtrado = df[df['median_house_value'] >= 150000]
filtrado
````

````python
##Gerar gráfico
import matplotlib.pyplot as plt
````

````python
# Criar o scatter plot com cores baseadas em 'median_house_value'
plt.figure(figsize=(10, 6))
scatter = plt.scatter(filtrado['longitude'], filtrado['latitude'],
                      c=filtrado['median_house_value'],  # Define as cores conforme o valor
                      cmap='viridis',                  # Escolhe um colormap
                      alpha=0.7)                       # Transparência dos pontos

# Adicionar uma barra de cores para referência
cbar = plt.colorbar(scatter)
cbar.set_label('Median House Value')

# Configurar rótulos e título
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Distribuição Geográfica - Casas com Valor Médio >= 150000')
plt.show()
````

````python
##Gerando gráfico
import geopandas as gpd
from shapely.geometry import Point
import contextily as ctx
````

````python
# Criar uma lista de geometria (pontos) a partir das colunas longitude e latitude
geometry = [Point(xy) for xy in zip(filtrado['longitude'], filtrado['latitude'])]

# Converter o DataFrame para um GeoDataFrame, definindo o CRS como EPSG:4326 (WGS84)
geo_df = gpd.GeoDataFrame(filtrado, geometry=geometry, crs="EPSG:4326")

# Converter para EPSG:3857 (Web Mercator), necessário para o contextily
geo_df = geo_df.to_crs(epsg=3857)

# Criar a figura e os eixos
fig, ax = plt.subplots(figsize=(10, 8))

# Plotar os pontos com cores baseadas em 'median_house_value'
scatter = ax.scatter(geo_df.geometry.x, geo_df.geometry.y,
                     c=geo_df['median_house_value'],
                     cmap='viridis',
                     alpha=0.7)

# Adicionar o basemap (você pode escolher outros provedores, ex: ctx.providers.Stamen.TonerLite)
ctx.add_basemap(ax, source=ctx.providers.CartoDB.Positron)

# Remover os eixos (opcional)
ax.set_axis_off()

# Adicionar a barra de cores
cbar = plt.colorbar(scatter, ax=ax, fraction=0.036, pad=0.04)
cbar.set_label('Median House Value')

# Título do gráfico
plt.title('Distribuição Geográfica - Casas com Valor Médio >= 150000')
plt.show()
````
